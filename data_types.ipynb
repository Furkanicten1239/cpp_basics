## C++ Notes: Data Types


```c++17
#include <iostream>
  	// Creating a variable to store integer
    int var = 68;
  
  	std::cout << var;

    return 0;
```

Data types specify the type of data that a variable can store. Whenever a variable is defined in C++, the compiler allocates some memory for that variable based on the data type with which it is declared as every data type requires a different amount of memory.



In the above code, we needed to store the value 68 in our program, so we created a variable var. But before var, we have used the keyword 'int'. This keyword is used to define that the variable var will store data of type integer.

**Integer (int):** Its size is 4-bytes (for 64-bit) systems and can store numbers for binary, octal, decimal and hexadecimal base systems in the range from -2,147,483,648 to 2,147,483,647/

**Character (char):** Used to store a single character. Its size is 1 byte. It can generally store upto 256 characters according to their ASCII codes.


```c++17
  	char c = 'A';  //Numbers can also be stored as characters.
	std::cout << c;

    return 0;
```

**Booelan (bool):** Used to store true-false. It is 1 byte

```c++17
#include <iostream>

void printBools() {
    bool isTrue = true;
    bool isFalse = false;
    std::cout << "isTrue: " << (isTrue ? "true" : "false") << "\n";
    std::cout << "isFalse: " << (isFalse ? "true" : "false") << "\n";
}

printBools();
```

**Float (double):** used to store numbers with decimal points. The keyword used to define floating-point numbers is float. Its size is 4 bytes (on 64-bit systems) and can store values in the range from 1.2e-38 to 3.4e+38.

```c++17
#include <iostream>

int main() {
    float f = 36.5;
    std::cout << f << "\n";
    return 0;
}

main();  // call it from xeus-cling
```

**Double (double):** It also defines decimal point numbers but with 8byte and from 1.7e-308 to 1.7e+308

```c++17
#include <cstdio> // for printf

int main() {
    double pi = 3.141592653589793;
    std::printf("%.15f\n", pi);
    return 0;
}

main();
```

**Data Type Conversion:** 
- Type conversion refers to the process of changing one data type into another compatible one without losing its original meaning. It's an important concept for handling different data types in C++.
- %d  print an integer (decimal).
- %c  print a character.
- \n  newline.

```c++17
#include <iostream>

int main() {
    int n = 3;
    char  c = 'C';
    
    // Convert char data type into integer
    std::printf("%c as integer:" " %d\n", c, int(c));
    int sum = n + c;
    std::printf("Sum of %d and %c (as integer %d) is: %d\n", n, c, int(c), sum);
    return 0;
}
main();  // call it from xeus-cling
```

size of the variables can be learned by using sizeof() command. Moreover, every variable size can be expanded by adding "long".

```c++17
#include <iostream>

int main() {

    // Printing the size of each data type
    // %zu is the format specifier for size_t.
    std::printf("Size of %s : %zu bytes \n", "int", sizeof(int));
    std::printf("Size of %s : %zu bytes \n", "char", sizeof(char));
    std::printf("Size of %s : %zu bytes \n", "float", sizeof(float));
    std::printf("Size of %s : %zu bytes \n", "double", sizeof(double));
    std::printf("Size of %s : %zu bytes \n", "long double", sizeof(long double));

    return 0;
}
main();
```
