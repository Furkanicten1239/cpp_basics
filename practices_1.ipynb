# FIRST QUESTION  
Now write the function with the prototype bool isPrime(int n); to check if the input n is prime.
Consider the following approach to check if a number is prime:
- A prime number has no divisors except 1 and itself.
- It is enough to check divisors only up to âˆšn.
- Use the cmath library and the sqrt() function.


```c++
#include <iostream>
#include <cmath>

// --- Your Function Definition ---
// I've also corrected the logic and the return type.
void is_prime() {
    int number;
    std::cout << "Enter a number: "; // Prompt the user
    std::cin >> number;
    std::cout << "You chose the number: " << number << std::endl;

    bool isprime = true;

    // 1. Handle base cases: Numbers 1 and less are not prime.
    if (number <= 1) {
        isprime = false;
    } else {
        // 2. Loop to check for divisors.
        for (int i = 2; i <= sqrt(number); i++) {
            if (number % i == 0) {
                isprime = false;
                break; // Exit the loop immediately once a divisor is found.
            }
        }
    }
    
    // 3. Print the final result.
    if (isprime) {
        std::cout << number << " is prime." << std::endl;
    } else {
        std::cout << number << " is not prime." << std::endl;
    }
    // A 'void' function does not return a value.
}
```

# SECOND QUESTION

Write a function with the prototype bool isSquare(int n); to check if an integer is a perfectsquare.
- Use sqrt() from cmath.
- Return true if the number is a square, otherwise false.


```c++
#include <iostream>
#include <cmath>
bool isSquare(int n){

    int int_sqrt;
    int_sqrt = sqrt(n);

    // We used that print methods to check the output.
    //cout << sqrt(n) << endl;
    //cout << int_sqrt << endl;
    if (n < 0) return false;
    return (sqrt(n) == int_sqrt);
}
```

lets test the code

```c++
int main(){
    int n = 8;
    std::cout << "is number square ? " << isSquare(n) << std::endl;
    return 0;
}
main();
```

Or you can directly check their multiplication is equal to the integer n or not.

```c++
#include <iostream>
#include <cmath>
bool isSquare2(int n){
    int sqrt_n = sqrt(n);
    if (n < 0) return false;
    return (sqrt_n * sqrt(n) == n);
}
```
